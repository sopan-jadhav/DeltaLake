##Schema Enforcement##

```python
data = [(1,100,"aa", 11),(2,200,"bb",22),(3,300,"cc",23),(4,400,"dd",24)]
df = spark.createDataFrame(data, ["id","salary","name","sus_cl"])
```

```python
df.show()
```

```python
df.write.format("delta")\
    .mode("append")\
    .save("/Volumes/workspace/default/deltavol/demosink/")
```


##Schema Evolution##


```python
df.write.format("delta")\
    .mode("append")\
    .option("mergeSchema", "true")\
    .save("/Volumes/workspace/default/deltavol/demosink/")
```
