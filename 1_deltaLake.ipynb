## GENERATED COLUMNS ##

## Identity Columns ##

```python
from delta.tables import DeltaTable, IdentityGenerator
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
DeltaTable.create(spark)\
    .tableName("deltalakesopan.default.firsttableapi")\
    .addColumn("id_col1", dataType=LongType(), generatedAlwaysAs=IdentityGenerator())\
    .addColumn("salary", dataType = IntegerType())\
    .addColumn("name", dataType = StringType())\
    .execute()
```

## Computed Columns ##

```python
DeltaTable.create(spark)\
    .tableName("deltalakesopan.default.firsttableapi_1")\
    .addColumn("SalaryAfterTax", dataType=LongType(), generatedAlwaysAs="Cast((Salary * 0.7) as BIGINT)")\
    .addColumn("salary", dataType = IntegerType())\
    .addColumn("name", dataType = StringType())\
    .execute()
```

```python
%sql
insert into deltalakesopan.default.firsttableapi_1(salary, name)
values
(100,'aa'),
(200, 'bb')
```
