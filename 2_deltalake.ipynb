##Delta Logs##

```python
data = [(1,100,"aa"),(2,200,"bb"),(3,300,"cc"),(4,400,"dd")]
df = spark.createDataFrame(data, ["id","salary","name"])
```

```python
df.show()
```

```python
# Here I want to insert the data so I'm using append
df.write.format("delta")\
    .mode("append")\
    .save("/Volumes/workspace/default/deltavol/demosink/")
```
