##DML, Delete, Update##

```python
data = [(1,100,"aa", 11),(2,200,"bb",22),(3,300,"cc",23),(4,400,"dd",24)]
df = spark.createDataFrame(data, ["id","salary","name","sus_cl"])
df.write.format("delta")\
    .mode("append")\
    .save("/Volumes/workspace/default/deltavol/dmlsink/")
```

```python
%sql
select * from delta.`/Volumes/workspace/default/deltavol/dmlsink/`
```

##Update##

```python
%sql
update delta.`/Volumes/workspace/default/deltavol/dmlsink/`
set salary = 10000 where id = 1
```
